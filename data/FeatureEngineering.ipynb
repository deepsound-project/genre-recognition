```python
import pandas as pd
import numpy as np
```

```python
raw_track_features = pd.read_csv('../../fma_metadata/tracks.csv')
raw_track_genres = pd.read_csv('../../fma_metadata/genres.csv')
```

```python
columns = raw_track_features.loc[0, :].values
columns[0] = 'track_id'
df_values = raw_track_features.loc[2:, :].values
```

```python
track_features = pd.DataFrame(df_values.T, columns).T
parent_genres = raw_track_genres[raw_track_genres['parent'] == 0]
parent_genres_dict = create_genre_dict(parent_genres)
track_features['parent_genre'] = track_features['genre_top'].map(parent_genres_dict)
id_to_genre_dict = pd.DataFrame(np.array([track_features['track_id'].values, track_features['parent_genre'].values]).T, columns=['track_id', 'genre_id'])
id_to_genre_dict.dropna()
```

```python
def create_genre_dict(genres):
    return dict(zip(genres['title'], genres['genre_id']))
```

```python
id_to_genre_dict.to_csv('track_to_genre.csv')
```
